```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
daily_visitors = pd.read_csv("VenueDailyVisitors.csv", index_col = 0)
age = pd.read_csv("VenueAge.csv", index_col = 0)
distance = pd.read_csv("VenueDistance.csv", index_col = 0)
duration = pd.read_csv("VenueDuration.csv", index_col = 0)
gender = pd.read_csv("VenueGender.csv", index_col = 0)
spend = pd.read_csv("VenueSpend.csv", index_col = 0)
```

```python
data = pd.DataFrame(daily_visitors.mean(), columns = ['Average daily visitors'], index = age.index)
```

```python
data['Age (years)'] = age.values
data['Distance (miles)'] = distance.values
data['Duration (mins)'] = duration.values
data['Gender (female proportion %)'] = gender.values
data['Spend (Â£)'] = spend.values
```

```python
plt.figure(figsize=(12, 12))
corr = data.corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap="PiYG", square=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_title('Heat map', fontsize=30)
plt.show()
plt.savefig("Visualisation 1 - Heat map.png")
```
